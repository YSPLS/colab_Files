Basic Statistics

```python
df.head()
```

```python
print(df['Price'].dtype)

#Dataan kerätyt tiedot ovat intialaiselta autokauppasivustolta, joten rahayksikköinä on Intian rupia.
#Intiassa Lakh on lukusana, joka tarkoittaa 100,000.
```

```python
average = np.average(df['Price']) # Kaikkien Käytettyjen autojen hintojen keskiarvo
print("The average price of all the used cars is: ", average, "Lakh INR")
```

```python
median = np.median(df['Price']) #Kaikkien käytettyjen autojen hintojen mediaani.
print("The median price of all the used cars is: ", median, "Lakh INR")
```

```python
std = np.std(df['Price']) # Kaikkien käytettyjen autojen hintojen keskihajonta
print("The standard deviation of all the used car prices is: ", std, "Lakh INR")
```

```python
var = np.var(df['Price']) # Kaikkien käytettyjen autojen hinnan varianssi
print("The variance of all the used car prices is: ", var, "Lakh INR")
```

```python
max = np.max(df['Price']) # Kaikkien käytettyjen autojen maksimi hinta
print("The maximum of all the used car prices is: ", max, "Lakh INR")
```

```python
min = np.min(df['Price']) # Kaikkien käytettyjen autojen minimi hinta
print("The minimum of all the used car prices is: ", min, "Lakh INR")
```

```python
sum = np.sum(df['Price']) # Kaikkien käytettyjen autojen hintojen summa
print("The sum of all the used car prices is: ", sum, "Lakh INR")
```

Visualization

```python
# Line Plot
df.plot(x='Year', y='Kilometers_Driven', kind='line', title='Kilometers driven over years', marker='o', color='blue') #lisätään väri ja marker tyyli ja title
plt.xlabel('year') #label x-axista varten
plt.ylabel('Kilometers Driven') #label y-axista varten
plt.grid(True) #lisää grid viivat
plt.show()
```

https://imgur.com/a/7iwOahk, line plot kuva

```python
df['Mileage'] = df['Mileage'].str.extract('(\d+\.\d+|\d+)').astype(float)# muuttaa 'Mileage' floatiksi
sampled_data = df.sample(n=500) #pienentää datan kokoa että taulu näyttää järkevältä
#scatter plot
sampled_data.plot(x='Mileage', y='Price', kind='scatter', title='Mileage vs Price', color='green') #lisätään väri ja titlet ja mitä käytetään
plt.xlabel('Mileage')  #label x-axista varten
plt.ylabel('Price')  #label y-axista varten
plt.show()
```

https://imgur.com/a/hjTO1do, scatter plot kuva

```python
# Bar Chart
bar_data = df['Fuel_Type'].value_counts() #lisätään mitä dataa käytetään
bar_data.plot(kind='bar', title='eri polttoaine jakaumat', color=['blue', 'green', 'orange']) #lisätään värit ja title
plt.xlabel('polttoaine tyyppi')  #label x-axista varten
plt.ylabel('määrä')  #label y-axista varten
plt.show()
```

https://imgur.com/OwilFi0, bar chart kuva

```python
df['Power'] = df['Power'].str.extract('(\d+\.\d+|\d+)').astype(float) # muuttaa 'power' floatiksi
# Histogram
df['Power'].dropna().astype(float).hist(bins=20, edgecolor='black', color='purple',) #lisätään mitä ja moneen biniin se jakautuu ja värit
plt.title('Histogram of Power') #lisätään title
plt.xlabel('Power') #label x-axista varten
plt.ylabel('Frequency') #label y-axista varten
plt.show()
```

https://imgur.com/N2vVnll, histogrammin kuva

```python
# Pie Chart
pie_data = df['Transmission'].value_counts() #valitaan data
pie_data.plot(kind='pie', autopct='%1.1f%%', title='vaihde luokan jaukauma') #valitaan title ja miten data jakautuu
plt.show()
```

https://imgur.com/QklZmOz, pie chartin kuva

```python
df['Engine'] = df['Engine'].str.extract('(\d+)').astype(float) # muuttaa 'Engine' floatiksi
# Heatmap (seaborn)
heatmap_data = df[['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Price']] #valitaan mitkä laitetaan heatmappiin
sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm') #lisätään numerot jokaseen ruutuun ja värjäys tapa
plt.title('yhteväisyydet') #lisätään title
plt.show()
```
